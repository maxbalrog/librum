## Pipeline demo

Given an image of single bookshelf cell, Hough lines are detected on Canny edge map. Dominant line orientation is defined through histogram of angles $\theta$. The image is rotated, so book spines appear in vertical position while Hough lines are filtered both in angle and distance from each other (close Hough lines are merged into one).

```python
idx = 0
path = dataset[idx]
test_img = cv2.imread(path)

print('Histogram of dominant Hough line angle before and after rotation:')
img_resc, img_rotated, lines_filtered, lines_rotated = rotate_img(test_img, plot_hist=True)
books = extract_books_warped(img_rotated, lines_rotated)

img1 = plot_lines(img_rotated, lines_rotated)
plot_img(img1)

plt.figure(dpi=150)
for i,book in enumerate(books):
    plt.subplot(1,len(books),i+1)
    plt.imshow(cv2.cvtColor(book, cv2.COLOR_BGR2RGB))
    plt.axis('off')
plt.show()
```

To detect the title, we additionally crop book image to the region around its title through histogram of bounding boxes for single contours. Then each book spine is windowed and Otsu thresholded (we suppose that book cover structure may differ significantly only in vertical direction, in horizontal there are mostly 2 colors: one for background and one for text). We additionally check that black goes for background and white for text, if this is not the case - the mask in inverted. Then the contours of overall mask image are filtered with some heuristics to reduce the number of false "letter contours". Sometimes it may delete letters because it's not feasible to take into account all variety of text fonts and title location on the book spine.

```python
idx = -1
book_img = books[idx]
book_img_cropped = find_title_region(book_img)

print('Slightly crop image aroung title region')
plt.figure(dpi=150)
for i,book in enumerate([book_img,book_img_cropped]):
    plt.subplot(1,2,i+1)
    plt.imshow(cv2.cvtColor(book, cv2.COLOR_BGR2RGB))
    plt.axis('off')
plt.show()

book_mask, mask, contours_filtered, windows = book2mask(book_img_cropped, return_windows=True)

print('Windowed Otsu thresholding')
plt.figure(dpi=150)
for i,(window,window_mask) in enumerate(windows):
    plt.subplot(len(windows),2,2*i+1)
    plt.imshow(cv2.cvtColor(window, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(len(windows),2,2*i+2)
    plt.imshow(window_mask, cmap='gray')
    plt.axis('off')
plt.show()

print('Filter contours')
plt.figure(dpi=150)
for i,book in enumerate([book_mask,mask]):
    plt.subplot(1,2,i+1)
    plt.imshow(cv2.cvtColor(book, cv2.COLOR_BGR2RGB))
    plt.axis('off')
plt.show()
```

To merge remaining contours into separate words we construct a path through all contours starting from the top. Contours at relatively low distance from each other are merged into one word (in this contex by distance between contours we mean $L_2$-norm between it's nearest pixels). Such approach will result in oversegmented words. To alleviate such unfortunate turn of events, statics (distance mean, std and mean angle) for separate words is collected and used to merge near lying contours. If words are horizontal - they are automatically flipped. As for vertical text orientation: if text goes up-down - the obtained orientation is normal, if text goes down-up - obtained orientation is down-up (this doesn't matter because the trained detector should be robust to horizontal and vertical flips).

```python
print('Constructing a path and merging nearest contours together')
words_mask, words_connected_mask = cnt2words(mask, book_img_cropped, contours_filtered, thr=2.,
                                             connect=True, verbose=True)

img1 = book_img_cropped.copy()
for word in words_connected_mask['cnt']:
    x,y,w,h = cv2.boundingRect(word)
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),3)

print('Detected "words"')
plot_img(img1)

for word_mask in words_connected_mask['mask']:
    word_mask = np.rot90(word_mask, k=2)
    plot_img(word_mask)
```
